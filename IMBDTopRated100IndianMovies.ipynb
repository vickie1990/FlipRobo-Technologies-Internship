```python
# # import BeautifulSoup
from bs4 import BeautifulSoup
import requests

# request IMDb page of top rated 100 Indian movies for data
page = requests.get('https://www.imdb.com/india/top-rated-indian-movies/')
page

# request IMDb page of top rated 100 Indian movies for data
src = BeautifulSoup(page.content)

# scrape data for movie titles and ranking
top = []
for i in src.find_all("td",class_="titleColumn"):
    top.append(i.text.replace('\n',''))

# scrape data for movie IMDb ratings
rating = []
for i in src.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text.replace('\n',''))

# displaying the data scraped from IMDb site for top 100 Indian movies
for val in range(0,100):
    print(top[val],'  { IMDb Rating:',rating[val],'}')

import pandas as pd

df = pd.DataFrame({ 'IMDb Rating:',rating[val],})
df
```
