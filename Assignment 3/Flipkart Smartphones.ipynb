Q.4 Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV

```python
import pandas as pd
import selenium
from selenium import webdriver
import time
from selenium.common.exceptions import NoSuchElementException
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
```

```python
driver=webdriver.Chrome(r'C:\Users\hp\Downloads\chromedriver_win32\chromedriver.exe')
```

```python
driver.get("https://www.flipkart.com")
```

```python
Search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
Search
```

```python
search_btn.click()
```

```python
Url=[]
for i in driver.find_elements(By.XPATH,"//a[@class='_1fQZEK']"):
    Url.append(i.get_attribute("href"))
```

```python
len(Url)
```

```python
Url
```

```python
Brand_Name=[]
Smartphone_name=[]
Colour=[]
RAM=[]
Storage_ROM=[]
Primary_Camera_and_Secondary_Camera=[]
Display_Size=[]
Battery_Capacity=[]
Price=[]
Product_URL=[]
```

```python
for i in Url:
    driver.get(i)
    time.sleep(10)
   #fetching Brand_Name
    try:
        name=driver.find_element(By.XPATH,'//span[@class="B_NuCI"]')
        Brand_Name.append(name.text.replace("/n","new line"))
    except:
        Brand_Name.append('-')
    #fetching Smartphone_name
    try:
        name=driver.find_element(By.XPATH,'//table[@class="_14cfVK"]/tbody/tr[3]/td[2]/ul/li')
        Smartphone_name.append(name.text.replace("/n","new line"))
    except:
        Smartphone_name.append('-')
        #fetching Colour 
    try:
        name=driver.find_element(By.XPATH,'//table[@class="_14cfVK"]/tbody/tr[4]/td[2]/ul/li')
        Colour.append(name.text.replace("/n","new line"))
    except:
        Colour.append('-')
        #fetching RAM   
    try:
        name=driver.find_element(By.XPATH,'//div[@class="flxcaE"]/table/tbody/tr[2]/td[2]/ul/li')
        RAM.append(name.text.replace("/n","new line"))
    except:
        RAM.append('-')
        #fetching Storage(ROM)       
    try:
        name=driver.find_element(By.XPATH,'//li[@class="_21Ahn-"][1]')
        Storage_ROM.append(name.text.replace("/n","new line"))
    except:
        Storage_ROM.append('-')    
        #fetching Primary_Camera & Secondary_camera
    try:
        name=driver.find_element(By.XPATH,'//li[@class="_21Ahn-"][3]')
        Primary_Camera_and_Secondary_Camera.append(name.text.replace("/n","new line"))
    except:
        Primary_Camera_and_Secondary_Camera.append('-')  
        #fetching Display_Size
    try:
        name=driver.find_element(By.XPATH,'//div[@class="_2418kt"]/ul/li[2]')
        Display_Size.append(name.text.replace("/n","new line"))
    except:
        Display_Size.append('-')  
        #fetching Battery Capacity
    try:
        name=driver.find_element(By.XPATH,'//div[@class="_2418kt"]/ul/li[4]')
        Battery_Capacity.append(name.text.replace("/n","new line"))
    except:
        Battery_Capacity.append('-')
        #fetching Price
    try:
        name=driver.find_element(By.XPATH,'//div[@class="_30jeq3 _16Jk6d"]')
        Price.append(name.text.replace("/n","new line"))
    except:
        Price.append('-')  
        #fetching Product URL
    try:
        name=driver.find_element(By.XPATH,'//div[@class="_1fQZEK"]')
        Product_URL.append(name.text.replace("/n","new line"))
    except:
        Product_URL.append('-')  
```

```python
len(Brand_Name),len(Price),len(Smartphone_name),len(Colour),len(RAM),len(Storage_ROM),len(Primary_Camera_and_Secondary_Camera),len(Display_Size),len(Product_URL),len(Battery_Capacity)
```

```python
Product=pd.DataFrame()
Product['Brand name']= Brand_Name
Product['Price']= Price
Product['Smartphone_name']= Smartphone_name
Product['Colour']= Colour
Product['RAM|Storage_ROM']= Storage_ROM
Product['Primary_Camera|Secondary_Camera']= Primary_Camera_and_Secondary_Camera
Product['Display_Size']= Display_Size
Product['Battery_Capacity']= Battery_Capacity
Product['Product_URL']= Url
Product
```
