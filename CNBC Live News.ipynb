```python
# Python program to display the Headline on CNBC Global Site with time and news link

# import BeautifulSoup
from bs4 import BeautifulSoup
import requests

# request CNBC homepage for data
page = requests.get('https://www.cnbc.com/world/?region=world')
page

# extract data from CNBC homepage
src = BeautifulSoup(page.content)

# scrape data for latest news headline
news = []
for i in src.find_all("a",class_="LatestNews-headline"):
    news.append(i.text.replace('\n',''))

# scrape data for time of headline
time = []
for i in src.find_all("time",class_="LatestNews-timestamp"):
    time.append(i.text.replace('\n',''))

# scrape data for news link
link = []
for i in src.find_all("a",class_="LatestNews-headline"):
    link.append(i['href'])
    
# display news | time | news link
for val in range(0,len(news)):
    print(val+1,news[val],'|',time[val],'\n',link[val],'\n')

```
