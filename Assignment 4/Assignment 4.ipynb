Q1) Scrape the details of most viewed videos on YouTube from Wikipedia.Url = https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videosYou need to find following details:
A) Rank
B) Name
C) Artist
D) Upload date
E) Views

```python
#First connect to the driver
driver=webdriver.Chrome(r"C:\Users\hp\Downloads\chromedriver_win32\chromedriver.exe")
driver
```

```python
#open the page in automated chrome browser
url='https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos'
driver.get(url)
```

```python
# Creating Empty list to store all the scrapped result
rank=[]
name=[]
artist=[]
upload_date=[]
views=[]
# Scraping rank
r=driver.find_elements(By.TAG_NAME,"tr")
for i in r[2:32]:
    rank.append(i.text.split("\n")[0].split(".")[0])
    name.append(i.text.split("\n")[0].split(".")[1].split("[")[0])
#scraping Artist
c=4
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    
    artist.append(a[c].text)
    c=c+6
#scraping views
c=5
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    
    views.append(a[c].text)
    c=c+6
#scraping upload_date
c=6
a=driver.find_elements(By.TAG_NAME,"td")
for i in range(30):
    upload_date.append(a[c].text)
    c=c+6

details=pd.DataFrame(list(zip(rank,name,artist,upload_date,views)),
                          columns=['Rank','Name','Artist','Upload Date','Views'])
details    
```

```python
#First connect to the driver
driver=webdriver.Chrome(r"C:\Users\hp\Downloads\chromedriver_win32\chromedriver.exe")
driver
```

```python
#open the page in automated chrome browser
url='https://www.imdb.com/list/ls095964455/'
driver.get(url)
```

```python
# Creating Empty list to store all the scrapped result
name=[]
year_span=[]
genre=[]
run_time=[]
ratings=[]
votes=[]
#Scraping Name 
a=driver.find_elements(By.XPATH,"//h3[@class='lister-item-header']")
for i in a:
    name.append(i.text.split(".")[1].split("(")[0])
    
#Scraping year of span
a=driver.find_elements(By.XPATH,"//span[@class='lister-item-year text-muted unbold']")
for i in a:
    year_span.append(i.text.split("(")[1].split(")")[0])
#Scraping genre
a=driver.find_elements(By.XPATH,"//span[@class='genre']")
for i in a:
    genre.append(i.text)
#Scraping run time
a=driver.find_elements(By.XPATH,"//span[@class='runtime']")
for i in a:
    run_time.append(i.text)
#Scraping rating
a=driver.find_elements(By.XPATH,"//span[@class='ipl-rating-star__rating']")
for i in a:
    if i.text=="\n" or i.text=="Rate" or i.text=="":
        continue
    ratings.append(i.text)  
    
##Scraping votes
a=driver.find_elements(By.XPATH,"//p[@class='text-muted text-small']")

for i in range(2,len(a),3):
    votes.append(a[i].text.split(":")[1])
    
show_details=pd.DataFrame(list(zip(name,year_span,genre,run_time,ratings,votes)),
                          columns=['Show Name','Year of Span','Genre','Run Time','Rating','Votes'])
show_details    
```
