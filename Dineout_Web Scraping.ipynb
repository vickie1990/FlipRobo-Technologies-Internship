```python
# import BeautifulSoup
from bs4 import BeautifulSoup
import requests

# request Dineout webpage for data
page = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page

# extract data from Dineout webpage
src = BeautifulSoup(page.content)

# scrape data for name of the resturant
name = []
for i in src.find_all("a",class_="restnt-name ellipsis"):
    name.append(i.text.replace('\n',''))

# scrape data for cuisine
cuisine = []
for i in src.find_all("div",class_="detail-info"):
    cuisine.append(i.text.replace('\n',''))

# scrape data for resturant location    
location = []
for i in src.find_all("div",class_="restnt-loc ellipsis"):
    location.append(i.text.replace('\n',''))

# scrape data for resturant rating
rating = []
for i in src.find_all("div",class_="restnt-rating rating-4"):
    rating.append(i.text.replace('\n',''))

# scrape data for resturant image URL
img = []
for i in src.find_all("img",class_="no-img"):
    img.append(i['data-src'])

# display the data collected. Resturant name | cuisine | location | rating | image URL 
for val in range(0,len(name)):
    print(val+1,name[val],'|',cuisine[val],'|',location[val],'| Rating:',rating[val],'|',img[val],'|\n')
    
# make a dataframe to store resturant information
import pandas as pd

db = pd.DataFrame({'Resturant Name':name,'Cuisine':cuisine,'Location':location,'Ratings':rating,'Image URL':img})
db.index +=1
db    
```
